```python
import pandas as pd

df = pd.read_csv("/content/smoke_detection_iot.csv")
df.info()
```

```python
df.head()
```

```python
df = df.dropna()
df = df.drop(df.columns[[0, 1, 14]], axis=1)
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Normalisasi
normalized_df = scaler.fit_transform(df)

# return hasil normalisasi ke dalam DataFrame
new_df = pd.DataFrame(normalized_df, columns=df.columns)

new_df.head()
```

```python
# split fitur (X) dan target (y)
X = new_df.drop(columns=['Fire Alarm'])
y = new_df['Fire Alarm']

X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
# Membagi dataset menjadi data train dan data tes
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1000)
```

```python
from sklearn.tree import DecisionTreeClassifier
# model decision tree classifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
# Evaluasi performa model dengan cross_val_score
scores = cross_val_score(clf, X, y, cv=5)
print("Cross-validation scores:", scores)
print("Mean cross-validation score:", scores.mean())
```

```python
# Contoh data untuk prediksi manual
manual_input = [[24, 55, 0, 444, 12345, 16789, 933, 0, 0, 0, 0, 0]]

from sklearn.preprocessing import StandardScaler

# Melakukan skala data
scaler = StandardScaler()
pengujian_scaled = scaler.fit_transform(manual_input)

# Melakukan prediksi
prediction = clf.predict(pengujian_scaled)
print("Hasil prediksi:", prediction)
#Fire Alarm: Menunjukkan apakah alarm kebakaran berbunyi atau tidak
```

```python

from sklearn.tree import export_graphviz
import graphviz

# Ekspor decision tree ke dalam format dot
dot_data = export_graphviz(clf, out_file=None, feature_names=X.columns,
                           class_names=['Non-Potable', 'Potable'], filled=True);

# Membuat objek grafik dari data dot
graph = graphviz.Source(dot_data);

# Menyimpan gambar pohon keputusan dalam file PDF
graph.render("decision_tree", format='pdf')
```
